```python
import os
import sys
import netCDF4
import xarray as xr
import numpy as np
import aostools.climate as ac
import matplotlib.pyplot as plt
```

```python
#FOLDER_PATH = "../data/"
files = f"{FOLDER_PATH}{os.sep}air.mon.mean.nc"
```

```python
fname = xr.open_dataset(files)
```

```python
fname
```

```python
level = fname.level
lat = fname.lat
lon = fname.lon
air = fname.air
lat
```

```python
air
```

```python
mon = air['time.month']==7
air_jul = air.where(mon,drop=True)
type(air_jul)
air_jul
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeat

fig, ax = plt.subplots(figsize=(16,10))
air_jul_mean = air_jul.mean('time')[1,:,:]
cs = air_jul_mean.plot.contourf( ax=ax,x = 'lat',y = 'lon',robust=True)
ax.set_xlim([0,90])
plt.show()
# ax,set_extent([-180, 180, 0, 90])
# save fig
# plt.savefig('~/Desktop/jcli/figures/1.png')
```
