# Background

Edited from [Wikipedia](https://en.wikipedia.org/wiki/Colorado_Mineral_Belt)

The Colorado Mineral Belt (CMB) is an area of ore deposits from the La Plata Mountains in Southwestern Colorado to near the middle of the state at Boulder, Colorado and from which over 25 million troy ounces (778 t) of gold were extracted beginning in 1858. The belt is a northeast-striking zone defined by: 

* A shear zone system aged 2500Ma to 541Ma.

* Anomalous concentrations of chalcophile and siderophilic elements.

* A major gravity low

* And abnormal heat flow considering its location over thick continental crust

The belt lies within a zone that has been geologically active at intervals beginning from at least 1.6 billion years ago until the present. Parts of the CMB follow shear zones of Precambrian through Mesozoic age (>541-65 Ma). Igneous rocks intruded about 60 to 70 million years ago during the Laramide orogeny are associated with the belt and once were thought to be responsible for most of the ore deposits. Now many of the important ore deposits are thought to be genetically related to younger magmatism, some as young as ~25 million years.

In this frame, Beds are only logged if they exist at a particular well. However, this does not capture that certain beds were not observed at some wells. We need to take this into consideration, and add 'sythetic' zero thickness data at every well where a particular well was not observed to have a bed. 

First, lets make a dictionary keyed by well names and containing empty extendable lists at every key

````python
wells = df['well #'].unique()
missing_wells = {}
for well in wells:
    missing_wells[well]=[]
    
missing_wells
````

Next, we need to loop over every unique bed_id and create dataframes which only contain that particular bed_id

````python
for bed_id in df['Bed ID'].unique():
    df_subset = df[df['Bed ID']==bed_id]
   
````

Using sets, we can find the difference between the set of all wells in our dataset and only those in the bed_id frames

````python
 for bed_id in df['Bed ID'].unique():
    df_subset = df[df['Bed ID']==bed_id]
   
    well_subset = df_subset['well #'].unique()
    wells_with_no_bed = set(wells).difference(well_subset)
````

Finally, we need to loop over these orphan wells and add them into our dictionary datastructure

````python
 for bed_id in df['Bed ID'].unique():
    df_subset = df[df['Bed ID']==bed_id]
   
    well_subset = df_subset['well #'].unique()
    wells_with_no_bed = set(wells).difference(well_subset)
    for missing_well in wells_with_no_bed:
        missing_wells[missing_well].append(bed_id)
        
missing_wells
````

We now need to extract coordinates for each well into a dictionary. 

````python
well_coordinates = {}
for well in wells:
    data_subset = df[df['well #']==well]
    x=data_subset.X.values[0]
    y=data_subset.Y.values[0]
    well_coordinates[well]=(x,y)
    
well_coordinates
````

we will create a list of dictionaries containing 'fake' 0 thickness data for every bed_id not present at a particular well 

````python
new_data = []
for well in missing_wells.keys():
    coordinate = well_coordinates[well]
    beds_to_add = missing_wells[well]
    for bed in beds_to_add:
        new_data.append({'X':coordinate[0],
                        'Y':coordinate[1],
                        'well #':well,
                        'Bed ID': bed,
                        'Thickness (m)': 0})
````

````python
new_data
````

````python
df_2 = df_1.append(new_data).reset_index().drop(columns='index')
df_2
````

If we want to make a contour map of geologic unit properties, it is clear that we don't have enough spatial coverage to accurately produce a simple interpolated map.

However, this is geology! We know that the thickness of layers doesn't change dramatically over distance. We also know that the closer in space we are to a measurement location, the more likely it is that the value we are interested in corresponds to the data there.

We have just described something called 'Krieging'. Originally developed as a way to estimate economic mineral grades across scattered data in mines, Krieging centeres a gaussian distribution of values on each datapoint, minimizing the total error of overlapping 'bell' curves. In machine learning, Krieging is a subset of something called Gaussian Processes with a radial basis function. With this in mind, lets deploy krieging on Bed ID 1 and see what happens

````python
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

print(df_2['Bed ID'].unique())
# This section create the grid of geo points to interpolate onto
x_min = df_2.X.min()
x_max = df_2.X.max()
y_min = df_2.Y.min()
y_max = df_2.Y.max()

x = np.linspace(x_min-100,x_max,num=1000)
y = np.linspace(y_min-100,y_max+100,num=1000)

X, Y = np.meshgrid(x, y)
# this section extracts the bed_id data to use for our Gaussian Process task
bed_id =1
target_bed = df_2[df_2['Bed ID']==bed_id]
print(target_bed['Thickness (m)'].values)

points = np.zeros((len(target_bed),2))
points[:,0]=target_bed.X.values
points[:,1]=target_bed.Y.values

values = target_bed['Thickness (m)'].values   # Replace with your observed data
# Then we create our gaussian process regressor using a krieging kernel, fitting it to our data
gp = GaussianProcessRegressor(kernel=RBF(length_scale=50),normalize_y=True)
gp.fit(points, values)

# finally we have our regressor try to predict the bed thickness over our interpolation points
XY_pairs = np.column_stack([X.flatten(), Y.flatten()])
predicted = gp.predict(XY_pairs).reshape(X.shape)
````

````python
fig, ax = plt.subplots(figsize=(10,10))
cs=plt.contourf(X,Y,predicted,cmap='gist_earth',levels=np.linspace(0,2,num=9))
cs=plt.contour(X,Y,predicted,colors='black',levels=np.linspace(0,2,num=9))
ax.clabel(cs, cs.levels, inline=True,fontsize=10)
plt.scatter(df_2.X,df_2.Y,color='white',zorder=4)
#plt.xlim([2700,3200])
#plt.ylim([200,500])
plt.show()
````
